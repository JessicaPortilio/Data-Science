# **Validação Cruzada**

```python
from sklearn.model_selection import cross_val_score, KFold
```

```python
resultados_random_forest = []
resultados_logistica = []
resultados_rede_neural = []
for i in range(30):
  print(i)
  kfold = KFold(n_splits=10, shuffle=True, random_state=i)


  forest = RandomForestClassifier(criterion='entropy', min_samples_leaf=1, min_samples_split=5, n_estimators=10)
  scores = cross_val_score(forest, X_train, y_train, cv = kfold)
  resultados_random_forest.append(scores.mean())

  
  logistic = LogisticRegression(C=1.0, solver='lbfgs', tol=0.0001)
  scores = cross_val_score(logistic, X_train, y_train, cv = kfold)
  resultados_logistica.append(scores.mean())

  

  rede_neural= MLPClassifier(activation='logistic', batch_size=56, solver='adam')
  scores = cross_val_score(rede_neural,X_train, y_train, cv = kfold)
  resultados_rede_neural.append(scores.mean())
```

```python
resultados = pd.DataFrame({'Random forest': resultados_random_forest,
                           'Logistica': resultados_logistica,
                           'Rede neural': resultados_rede_neural})
resultados
```
